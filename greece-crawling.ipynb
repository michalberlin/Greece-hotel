In the following function, I randomly choose user-agent, so that I can access the server each time with a different user and reduce the chance of blocks.

```python
def get_random_user_agent():
    return random.choice(user_agents)
```

```python
selenium_options = Options()
selenium_options.add_argument("--user-agent={}".format(get_random_user_agent()))
```

The function returns a Soup object. Building a webdriver using selenium's chrome constructor.
In addition, there's a use of selenium commands to open an entire page and scroll down to be able to scrape all the data in each hotel, while waiting a few seconds after each command in order to let it execute.

```python
def create_soup_from_url(url):
    try:
        driver = webdriver.Chrome(options = selenium_options)
        
        driver.get(url)
        # Wait for the page to load
        time.sleep(2)
        # Open the page window in his maximal size
        driver.maximize_window()
        time.sleep(2)
        # Scrolling the web page so the price toolbar would appear and i can fetch its tag
        driver.execute_script("window.scrollTo(0, 100)")
        time.sleep(3)
        # Saving the HTML file into the variable
        html_content = driver.page_source
    
        soup = bs(html_content, 'html.parser')
        return soup
    except:
        print('error :/')
```

Defining the columns in the data frame and creating it

```python
defined_columns =['Hotel name', 'Hotel price', 'Reviews count', 'Location rating', 'Cleanliness rating', 'Service rating',
                  'Value rating', 'Pool_numerical', 'Free_parking_numerical', 'free_wifi_numerical', 'fitness_center_numerical',
                  'AC numerical', 'minibar numerical', 'cable tv numerical', 'non smoking hotel numerical',
                  'baggage storage numerical', 'housekeeping numerical', 'safe numerical', 'airport numerical',
                  'taxi numerical', 'pets allowed numerical', 'private balcony numerical', 'ocean view numerical',
                  'spa numerical', 'number of restaurants', 'number of attractions', 'rating']

df = pd.DataFrame(columns=defined_columns)
```

This function gets a data frame as a parameter and fills it with all of the data about hotels.
When a page that contains hotels is fetched, all of the hotels' links are added into a list that is iterated later in order to add any hotel's data into the data frame.

```python
def iterate_over_pages(df):
    #
    raw_url = r'https://www.tripadvisor.com/Hotels-g189398-oa{}-a_ufe.true-Greece-Hotels.html'
    addition_temp = 30

    while addition_temp <= 3150 and df.shape[0] <= 3200:
        temp_url = raw_url.format(addition_temp)
        soup = create_soup_from_url(temp_url)
    
        hotels_container = soup.find("div", class_='EJmPL _T e bADWL _T PnJDh')
        hotels_list = soup.find_all("div", class_='NXAUb _T')
        href_list = []
    
        for hotel in hotels_list:
            hotel_card_container = hotel.find('div', class_='jsTLT K')
            a_tag = hotel.find('a')['href']
            link_to_append = 'https://www.tripadvisor.com' + a_tag
            href_list.append(link_to_append)

        for href in href_list:
            temp_hotel_data = hotel_crawling(href)
            time.sleep(5)
            result_dict = {column:data for column, data in zip(defined_columns, temp_hotel_data)}
            # Convert result_dict to a dataframe
            new_row_df = pd.DataFrame(result_dict, index=[0])
            # Concatenate existing dataframe and new_row_df
            df = pd.concat([df, new_row_df], ignore_index=True)
            print(result_dict)
            print('dataframes size', df.shape[0])
        
        addition_temp+=30
        
    return df
```

This function gets an hotel link and fetches its data from using tags from the HTML, and returns it as a list.

```python
def hotel_crawling(hotel_link):
    soup = create_soup_from_url(hotel_link)

    try:
        temp_hotel_name = soup.find("h1", class_="biGQs _P rRtyp")
        hotel_name = temp_hotel_name.text
    except:
        hotel_name = None

    try:
        temp_num_of_reviews = soup.find("span", class_="biGQs _P pZUbB KxBGd")
        reviews_count_string = temp_num_of_reviews.text
        reviews_splitted_string = reviews_count_string.split(" ")
        reviews_numerical = reviews_splitted_string[0]
        reviews_numerical = reviews_numerical.replace(',', '')
    except:
        reviews_numerical = None

    try:
        temp_general_rating = soup.find("span", class_="uwJeR P")
        rating = temp_general_rating.text
    except:
        rating = None

    try:
        temp_price_raw = soup.find("div", class_="mcvYL b")
        price_string = temp_price_raw.text.replace(',', '')
        price_numeric = int(price_string[1:])
    except:
        price_numeric = None

    try:
        rating_sub = soup.find_all("div", class_='WdWxQ')
    except:
        rating_sub = []

    try:
        temp_location_rating = rating_sub[0].find("span", class_='CzVMJ').text
    except:
        temp_location_rating = None

    try:
        temp_cleanliness_rating = rating_sub[1].find("span", class_='CzVMJ').text
    except:
        temp_cleanliness_rating = None

    try:
        temp_service_rating = rating_sub[2].find("span", class_='CzVMJ').text
    except:
        temp_service_rating = None

    try:
        temp_value_rating = rating_sub[3].find("span", class_='CzVMJ').text
    except:
        temp_value_rating = None
    
    property_amenities_container = soup.find_all("div", class_='ssr-init-26f')
    
    try:
        property_facilities_container = property_amenities_container[4]
    except IndexError:
        property_facilities_container = None

    property_classes = property_facilities_container.find_all("div", class_='OsCbb K') if property_facilities_container else []
    try:
        property_amenities_extracted = property_classes[0].find_all("div", class_='yplav f ME H3 _c') if property_classes else []
        property_amenities_list = [property_amenity.text for property_amenity in property_amenities_extracted]
        spa_numeric = 1 if "Spa" in property_amenities_list else 0
        pool_numeric = 1 if "Pool" or "Private pool" in property_amenities_list else 0
        free_parking_numeric = 1 if "Free parking" in property_amenities_list else 0
        free_wifi_numeric = 1 if "Free High Speed Internet (WiFi)" in property_amenities_list else 0
        fitness_center_numeric = 1 if "Fitness Center with Gym / Workout Room" in property_amenities_list else 0
        taxi_numeric = 1 if "Taxi" in property_amenities_list else 0
        baggage_storage_numeric = 1 if "Baggage storage" in property_amenities_list else 0
        airport_transport_numeric = 1 if "Airport transportation" in property_amenities_list else 0
        pets_allowed_numeric = 1 if "Pets Allowed ( Dog / Pet Friendly )" in property_amenities_list else 0
    except:
        property_amenities_list = []
        spa_numeric =  0
        pool_numeric =  0
        free_parking_numeric =  0
        free_wifi_numeric =  0
        fitness_center_numeric =  0
        taxi_numeric =  0
        baggage_storage_numeric =  0
        airport_transport_numeric =  0
        pets_allowed_numeric = 0


    try:
        room_features_extracted = property_classes[1].find_all("div", class_='yplav f ME H3 _c') if property_classes else []
        room_features_list = [room_feature.text for room_feature in room_features_extracted]
        safe_numeric = 1 if "Safe" in room_features_list else 0
        housekeeping_numeric = 1 if "Housekeeping" in room_features_list else 0
        air_conditioning_numeric = 1 if "Air conditioning" in room_features_list else 0
        minibar_numeric = 1 if "Minibar" in room_features_list else 0
        private_balcony_numeric = 1 if "Private balcony" in room_features_list else 0
        tv_numeric = 1 if "Flatscreen TV" in room_features_list else 0
    except:
        room_features_list = []
        safe_numeric = 0
        housekeeping_numeric =  0
        air_conditioning_numeric =  0
        minibar_numeric =  0
        private_balcony_numeric = 0
        tv_numeric =  0


    try:
        room_types_extracted = property_classes[2].find_all("div", class_='yplav f ME H3 _c') if property_classes else []
        room_types_list = [room_type.text for room_type in room_types_extracted]
        non_smoking_hotel_numeric = 1 if "Non-smoking rooms" in room_types_list else 0
        ocean_view_numeric = 1 if "Ocean view" in room_types_list else 0

    except:
        room_types_list = []
        non_smoking_hotel_numeric =  0
        ocean_view_numeric =  0


    try:
        temp_num_of_restaurants = soup.find("span", class_="iVKnd Bznmz")
        num_of_restaurants = int(temp_num_of_restaurants.text)
    except:
        num_of_restaurants = None

    try:
        temp_num_attractions = soup.find("span", class_="iVKnd rYxbA")
        num_attractions = int(temp_num_attractions.text)
    except:
        num_attractions = None

    hotel_data = [hotel_name, price_numeric, reviews_numerical, temp_location_rating, temp_cleanliness_rating, temp_service_rating,
                  temp_value_rating, pool_numeric, free_parking_numeric, free_wifi_numeric, fitness_center_numeric,
                  air_conditioning_numeric, minibar_numeric, tv_numeric, non_smoking_hotel_numeric, baggage_storage_numeric,
                  housekeeping_numeric, safe_numeric, airport_transport_numeric, taxi_numeric, pets_allowed_numeric,
                  private_balcony_numeric, ocean_view_numeric, spa_numeric, num_of_restaurants, num_attractions, rating]
    return hotel_data
```
